```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Import the required libraries

```python
dc_df = pd.read_csv('./dc-wikia-data.csv')
marvel_df = pd.read_csv('./marvel-wikia-data.csv')
```

Read the CSV files, each as a seperate Pandas Dataframe object

```python
dc_df.info()
```

Look at the columns for the DC dataframe

```python
dc_df.drop(['page_id', 'urlslug', 'GSM', 'FIRST APPEARANCE'], axis=1, inplace=True)
dc_df.head()
```

Dropping unwanted columns and displaying the first 5 rows of the dataframe

```python
marvel_df.info()
```

Look at the columns for the DC dataframe

```python
marvel_df.drop(['page_id', 'urlslug', 'GSM', 'FIRST APPEARANCE'], axis=1, inplace=True)
marvel_df.head()
```

Again, dropping unwanted columns and displaying the first 5 rows

```python
plt.figure(figsize=(9,4))
sns.countplot(x='ID',data=dc_df,hue='ALIGN',alpha=0.7)
```

DC has characters with a bad alignment that tend to have a secret identity while good characters tend to have a public identity. The two groups have roughly the same number of characters each.

```python
plt.figure(figsize=(9,4))
sns.countplot(x='ID',data=marvel_df,hue='ALIGN',alpha=0.7)
```

In contrast, Marvel has a high amount of bad characters that have a secretive identity when compared to good and neutral characters. Marvel also has about the same number of good and bad characters with a public identity.

```python
plt.figure(figsize=(9,4))
sns.violinplot(x='ALIGN',y='YEAR',data=dc_df)
```

```python
plt.figure(figsize=(9,4))
sns.violinplot(x='ALIGN',y='Year',data=marvel_df)
```

```python
plt.figure(figsize=(9,4))
sns.distplot(dc_df['APPEARANCES'],bins=100,color='red')
sns.distplot(marvel_df['APPEARANCES'],bins=100,color='blue')
plt.legend(('DC','Marvel'))
```

```python
plt.figure(figsize=(9,4))
plt.xlim(right=100)
sns.distplot(dc_df['APPEARANCES'],bins=1000,color='red')
sns.distplot(marvel_df['APPEARANCES'],bins=1000,color='blue')
plt.legend(('DC','Marvel'))
```

```python
plt.figure(figsize=(16,4))
sns.countplot(x='EYE',data=dc_df,order=dc_df['EYE'].value_counts().iloc[:7].index)
plt.tight_layout()
```

```python
plt.figure(figsize=(16,4))
sns.countplot(x='EYE',data=marvel_df,order=marvel_df['EYE'].value_counts().iloc[:7].index)
plt.tight_layout()
```
